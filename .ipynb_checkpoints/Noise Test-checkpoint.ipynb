```python
import sys
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime
from pykalman import KalmanFilter
from statsmodels.nonparametric.smoothers_lowess import lowess
```

```python
df1 = pd.read_csv("WalkDrop1.csv",sep=';')
df2 = pd.read_csv("WalkDrop2.csv",sep=';')
df3 = pd.read_csv("WalkDrop3.csv",sep=';')
df4 = pd.read_csv("WalkSit.csv",sep=';')
df5 = pd.read_csv("WalkStairsSit.csv",sep=';')
df6 = pd.read_csv("WalkDrop.csv",sep=';')
df7 = pd.read_csv("WalkDropRedo.csv",sep=';')
df8 = pd.read_csv("JoshWalking.csv.csv",sep=';')
```

```python
filtered  = lowess(df1['AccelerationY'], df1['AccelerationX'] , frac=0.05)
```

```python
plt.plot(filtered[:, 0])
plt.plot(df1["AccelerationY"])
plt.show()
```

```python
plt.plot(filtered[:, 1])
plt.plot(df1["AccelerationX"])
plt.show()
```

```python
from scipy import signal
b, a = signal.butter(3, 0.1, btype='lowpass', analog=False)
low_passed = signal.filtfilt(b, a, df1["AccelerationX"])
```

```python
plt.plot(low_passed)
```

```python
low_passedy = signal.filtfilt(b, a, df1["AccelerationY"])
low_passedz = signal.filtfilt(b, a, df1["AccelerationZ"])
```

```python
plt.plot(low_passedy)
plt.show()
plt.plot(low_passedz)
```
